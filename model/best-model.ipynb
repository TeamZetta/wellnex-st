```python
#imported Kfold
from sklearn.model_selection import KFold
```

```python
## Function to run multiple algorithms with different K values of KFold.
def evaluate(train_data,kmax,algo):
    test_scores = {}
    train_scores = {}
    for i in range(2,kmax,2):
        kf = KFold(n_splits = i)
        sum_train = 0
        sum_test = 0
        data = df
        for train,test in kf.split(data):
            train_data = data.iloc[train,:]
            test_data = data.iloc[test,:]
            x_train = train_data.drop(["prognosis"],axis=1)
            y_train = train_data['prognosis']
            x_test = test_data.drop(["prognosis"],axis=1)
            y_test = test_data["prognosis"]
            algo_model = algo.fit(x_train,y_train)
            sum_train += algo_model.score(x_train,y_train)
            y_pred = algo_model.predict(x_test)
            sum_test += accuracy_score(y_test,y_pred)
        average_test = sum_test/i
        average_train = sum_train/i
        test_scores[i] = average_test
        train_scores[i] = average_train
        print("kvalue: ",i)
    return(train_scores,test_scores)   
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import nltk
import json

prognosis = json.load(open('./prognosis.json'))
VOCAB_DB = [' '.join(x.split('_')) for x in prognosis]
# print(VOCAB_DB)
MAX_NGRAM_LEN = max([len(x.split('_')) for x in VOCAB_DB])
print(MAX_NGRAM_LEN)

def vectorize(symptoms):
    found_symptoms = [int(x in symptoms) for x in prognosis]
    return found_symptoms

def extract_terms(text):
    tokens = nltk.word_tokenize(text)
    filtered_tokens = [word for word in tokens if word.isalpha()]
    vectorizer = CountVectorizer(ngram_range=(1, MAX_NGRAM_LEN), vocabulary=VOCAB_DB)
    filtered_str = ' '.join(filtered_tokens)
    print(filtered_str)
    X = vectorizer.transform([filtered_str])
    clf = MultinomialNB()
    clf.fit(X, ['health-related'])
    prediction = clf.predict(X)
    if prediction[0] == 'health-related':
        return [term.replace(' ','_') for term in vectorizer.get_feature_names_out() if term in filtered_str]
    return []


# natural_text = "I'm feeling fatigue, with loss of appetite, yellowing of eyes, sometimes yellowish skin. I have a family history."
natural_text = "I've been experiencing high temperature, severe headache, high fever, and fatigue for the past few days."
et = extract_terms(natural_text)
v = vectorize(et)
print(et)
print(v)
d_pred = dt.predict([v])
d_pred
```
